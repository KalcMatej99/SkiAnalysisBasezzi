```python
import pandas as pd
import numpy as np
```

```python
df_filtered = pd.read_csv("../data/filtered/Dati_filtrati_all.csv")
```

```python
for day in ["G1"]:#, "P1", "P2"]:
    df_filtered_day = df_filtered[df_filtered["Day"] == day]
    specific_edge_changes = pd.read_csv(f"../data/{day}/edge_changes/specific_edge_changes.csv")
    specific_edge_changes["change"] = range(1, len(specific_edge_changes) + 1)
    specific_edge_changes.set_index("change", inplace=True)
    specific_edge_changes = specific_edge_changes.transpose()
    for athlete, row in zip(specific_edge_changes.index, specific_edge_changes.to_numpy()):

        possible_rows = [row for row in df_filtered_day.to_numpy() if athlete in row[0]]

        if len(possible_rows) > 1:
            print(f"Multiple runs found for athlete {athlete} in day {day}")
        else:
            gps_data = pd.read_csv(possible_rows[0][1])
            
            start_time = gps_data.loc[0, "Timestamp"]

            longitudes, latitudes = [], []

            print("start", start_time)
            for change_time in row:
                change_time_timestamp = start_time + change_time
                #print(change_time_timestamp)
                gps_data["diff"] = np.abs(gps_data["Timestamp"] - change_time_timestamp)
                best_edge_change = gps_data[gps_data["diff"]==gps_data["diff"].min()]
                longitudes.append(best_edge_change["GNSS_Longitude"].to_numpy()[0])
                latitudes.append(best_edge_change["GNSS_Latitude"].to_numpy()[0])
            data = np.array([list(range(1, len(longitudes) + 1)), longitudes, latitudes])

            df_changes = pd.DataFrame(data.transpose(), \
                                      columns = ["num", "longitude","latitude"])
            print(df_changes)
            df_changes.to_csv(f"../data/G1/edge_changes/{possible_rows[0][0]}.csv", index=False)
```
